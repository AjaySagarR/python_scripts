```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from pathlib import Path


source_files = sorted(Path('new/').glob('*.csv'))
dataframes = []
for file in source_files:
    df = pd.read_csv(file)
    df = df.drop(columns=['D:0', 'D:1', 'D:2', 'phiOld', 'Normals:0', 'Normals:1', 'Normals:2','Points:2'])
    df.columns = ['x','y']
    df['r']=np.sqrt(df['x']**2 + df['y']**2)
    #df['source'] = file.name
    dataframes.append(df)

    
files=np.string_(source_files)
names_file = []
for i in files:
    names_file.append((int(str(i).split('.')[1]))*500)
    

data = pd.concat(dataframes,axis=1,keys = names_file)
data
```

```python
CbyA=[]
for i in names_file:
    ap=np.max(data[i]['r'])/np.min(data[i]['r'])
    CbyA.append(ap)
    
SF=(np.max(data[26000]['r']) - np.min(data[26000]['r']))/(np.max(data[26000]['r']) + np.min(data[26000]['r']))

SF
```
