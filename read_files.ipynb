```python
import pandas as pd
from pathlib import Path
import numpy as np

source_files = sorted(Path('').glob('*.csv'))
dataframes = []
for file in source_files:
    df = pd.read_csv(file)
    df = df.drop(columns=['D:0', 'D:1', 'D:2', 'phiOld', 'Normals:0', 'Normals:1', 'Normals:2','Points:2'])
    df.columns = ['x','y']
    #df['source'] = file.name
    dataframes.append(df)

data = pd.concat(dataframes,axis=1,keys = source_files)
```
