## Imports
  * Importing pandas, numpy, matplotlib, seaborn and other sklearn tools

```python
import numpy as np, seaborn as sns, pandas as pd, matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import math
```

## Get the Data
* Read in the finaldata.csv file as a DataFrame called df.

* finaldata.csv: Contains the data of radial distance at every 5 degree and corresponding Epsilon and Gamma values

```python
data = pd.DataFrame(pd.read_csv('finaldata.csv'))
data.head()
```

## Training and Testing Data

* Now we split the data into training and testing sets.
* Setting a variable X equal to the numerical features of the data ( E , Gamma ) and a variable y equal to the radial distance 
* Using **model_selection.train_test_split** from sklearn to split the data into training and testing sets. 
* Setting **test_size=0.3** which divides total data into training and testing data in the ratio of 70% and 30% respectively  
* For training the data, we are use **LinearRegression model**, using this we create a model named lm and then fit our training data in this model
* After this we predict the test data using this trained model

```python
X= data[['E' , 'Gamma']]
y =data[['Theta0', 'Theta5', 'Theta10', 'Theta15', 'Theta20','Theta25', 'Theta30', 'Theta35', 'Theta40', 'Theta45', 'Theta50','Theta55', 'Theta60', 'Theta65', 'Theta70', 'Theta75', 'Theta80','Theta85', 'Theta90', 'Theta95', 'Theta100', 'Theta105', 'Theta110','Theta115', 'Theta120', 'Theta125', 'Theta130', 'Theta135', 'Theta140','Theta145', 'Theta150', 'Theta155', 'Theta160', 'Theta165', 'Theta170','Theta175', 'Theta180', 'Theta185', 'Theta190', 'Theta195', 'Theta200','Theta205', 'Theta210', 'Theta215', 'Theta220', 'Theta225', 'Theta230','Theta235', 'Theta240', 'Theta245', 'Theta250', 'Theta255', 'Theta260','Theta265', 'Theta270', 'Theta275', 'Theta280', 'Theta285', 'Theta290','Theta295', 'Theta300', 'Theta305', 'Theta310', 'Theta315', 'Theta320','Theta325', 'Theta330', 'Theta335', 'Theta340', 'Theta345', 'Theta350','Theta355','Theta360']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
lm = LinearRegression()
lm.fit(X_train,y_train)
predictions = lm.predict(X_test)
```

## Evaluating the Model
* Comparing predicted values with test data 
* Calculating the Mean Absolute Error, Mean Squared Error, and the Root Mean Squared Error and R2 Score 

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
print('R2 SCORE:' , metrics.r2_score(y_test,predictions))
```

```python
%matplotlib inline
plt.figure(figsize=(6,6))
plt.scatter(y_test,predictions)
plt.xlabel('y_test', fontsize = 18)
plt.ylabel('predictions',fontsize = 18)
plt.title('Model Performance')
plt.show()
```
